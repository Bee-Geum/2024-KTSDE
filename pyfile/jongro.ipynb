```python
import osmnx as ox
```

```python
place_name = "종로구, 서울특별시, 대한민국"

graph = ox.graph_from_place(place_name, network_type='drive')
```

```python
fig, ax = ox.plot_graph(graph, node_color='black',edge_alpha=1, edge_color='black',\
                        bgcolor='white', node_size=0, edge_linewidth=0.2, figsize=(10, 10))
fig.savefig(f'./종로구.pdf')
```
